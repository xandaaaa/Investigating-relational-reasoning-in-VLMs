```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_grouped_bar_chart(accuracy_df, query_types=None):
    """Plot grouped bar chart for query types, averaged across all labels
    
    Args:
        accuracy_df: DataFrame with overall accuracy data
        query_types: List of query types to plot. If None, uses all query types in data.
                     Example: ['count', 'recognition_shape', 'explicit_connection']
    """
    
    # If no specific query types provided, use all from the data
    if query_types is None:
        query_types = accuracy_df['query_type'].unique().tolist()
    
    # Filter data for specified query types
    filtered_data = accuracy_df[accuracy_df['query_type'].isin(query_types)]
    
    
    # Average across all labels for each query_type
    plot_df = filtered_data.groupby('query_type').agg({
        'acc_all': 'mean',
        'acc_right': 'mean',
        'acc_false': 'mean'
    }).reset_index()
    
    # Define desired order for query types
    desired_order = [
        'count',
        'recognition_shape',
        'recognition_color',
        'recognition_shape_and_color',
        'explicit_connection',
        'explicit_arrow_direction',
        'implicit_spatial'
    ]
    
    # Create a categorical type with the desired order
    plot_df['query_type'] = pd.Categorical(
        plot_df['query_type'], 
        categories=desired_order, 
        ordered=True
    )
    
    # Sort by the categorical order
    plot_df = plot_df.sort_values('query_type').reset_index(drop=True)
    
    # Prepare data for plotting
    query_type_names = plot_df['query_type'].values
    acc_all = plot_df['acc_all'].values
    acc_right = plot_df['acc_right'].values
    acc_false = plot_df['acc_false'].values

    # Set up the bar chart
    x = np.arange(len(query_type_names))
    width = 0.25

    fig, ax = plt.subplots(figsize=(14, 7))

    # Create bars
    bars1 = ax.bar(x - width, acc_all, width, label='Overall Accuracy', color='#4472C4')
    bars2 = ax.bar(x, acc_right, width, label='Correct Answer Accuracy', color='#70AD47')
    bars3 = ax.bar(x + width, acc_false, width, label='False Answer Accuracy', color='#ED7D31')

    # sort the X labels based on counting, shape, color, shape_and_color, explicit connection, explicit arrow_direction, implicit spatial



    # Customize the chart
    ax.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
    ax.set_title('VLM Evaluation Performance by Query Type', 
                 fontsize=14, fontweight='bold', pad=20)
    ax.set_xticks(x)
    ax.set_xticklabels(query_type_names, rotation=45, ha='right', fontsize=10)
    ax.legend(loc='upper right', fontsize=10)
    ax.set_ylim(0, 1.05)
    ax.grid(axis='y', alpha=0.3, linestyle='--')

    # Add value labels on bars
    def autolabel(bars):
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        fontsize=7)

    autolabel(bars1)
    autolabel(bars2)
    autolabel(bars3)

    plt.tight_layout()
    plt.show()

# =========================
# PART 2: Line Plots - Accuracy across layers (Grid Layout)
# =========================

def plot_layer_accuracy_grid(accuracy_df_per_layer, query_types=None):
    """Plot accuracy across layers for query types in a grid layout
    Averages across all labels for each query type.
    
    Args:
        accuracy_df_per_layer: DataFrame with layer-wise accuracy data
        query_types: List of query types to plot. If None, uses all query types in data.
                     Example: ['count', 'recognition_shape', 'explicit_connection']
    """
    
    # If no specific query types provided, use all from the data
    if query_types is None:
        query_types = accuracy_df_per_layer['query_type'].unique().tolist()
    
    # Filter data for specified query types
    filtered_data = accuracy_df_per_layer[accuracy_df_per_layer['query_type'].isin(query_types)]
    
    if len(filtered_data) == 0:
        print("No data to plot")
        return
    
    # Average across all labels for each query_type and layer
    avg_data = filtered_data.groupby(['query_type', 'layer']).agg({
        'acc_all': 'mean',
        'acc_right': 'mean',
        'acc_false': 'mean'
    }).reset_index()
    
    # Get unique query types (these will be our subplots)
    combinations = [(qt,) for qt in avg_data['query_type'].unique()]
    
    # Calculate grid dimensions
    n_plots = len(combinations)
    n_cols = 3
    n_rows = (n_plots + n_cols - 1) // n_cols
    
    # Create figure with subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))

    # Flatten axes array for easier indexing
    if n_plots == 1:
        axes = np.array([axes])
    elif n_rows == 1 and n_cols == 1:
        axes = np.array([axes])
    elif n_rows == 1:
        axes = np.array(axes)
    else:
        axes = axes.flatten()
    
    # Plot each combination
    for idx, (query_type,) in enumerate(combinations):
        # Filter data for this query type
        data = avg_data[avg_data['query_type'] == query_type].sort_values('layer')
        
        if len(data) == 0:
            print(f"No data for {query_type}")
            continue
        
        ax = axes[idx]
        
        # Plot lines
        ax.plot(data['layer'], data['acc_all'], marker='o', linewidth=2, 
                label='All Inference', color='#2E86AB', markersize=6)
        
        if data['acc_right'].notna().any():
            ax.plot(data['layer'], data['acc_right'], marker='s', linewidth=2, 
                    label='Right Inference', color='#06A77D', markersize=6)
        
        if data['acc_false'].notna().any():
            ax.plot(data['layer'], data['acc_false'], marker='^', linewidth=2, 
                    label='False Inference', color='#D62246', markersize=6)
        
        ax.set_xlabel('Layer', fontsize=10)
        ax.set_ylabel('Accuracy', fontsize=10)
        ax.set_title(f'{query_type}', fontsize=11, fontweight='bold')
        ax.legend(fontsize=8, loc='best')
        ax.grid(True, alpha=0.3)
        ax.set_ylim([0, 1.05])
        
        # Set x-axis to show integer layers
        if len(data) > 0:
            ax.set_xticks(data['layer'].unique())

        # set x-axis only appear per 5 layers
        ax.set_xticks(np.arange(data['layer'].min(), data['layer'].max()+1, 5))
    
    # Hide unused subplots
    for idx in range(n_plots, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()


# =========================
# USAGE EXAMPLES
# =========================

# Example 1: Plot grouped bar chart (assuming accuracy_df exists)
plot_grouped_bar_chart(accuracy_df)

# Example 2: Plot layer accuracy grid - AVERAGED across all labels per query type
plot_layer_accuracy_grid(accuracy_df_per_layer, query_types=['explicit_connection', 'explicit_arrow_direction', 'implicit_spatial'])


print("Functions ready to use:")
print("1. plot_grouped_bar_chart(accuracy_df)")
print("2. plot_layer_accuracy_grid(accuracy_df_per_layer, query_types=None)")
print("   - Averages across ALL labels for each query type")
print("   - Pass query_types=['type1', 'type2'] to filter by query type")
print("   - Leave query_types=None to plot all query types")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_layer_accuracy_grid_right(accuracy_df_per_layer, query_types=None, figsize=(10, 6)):
    """
    Plot acc_right vs layer, with one line per query_type.
    Data is averaged across labels for each (query_type, layer).
    """

    # If no specific query types provided, use all from the data
    if query_types is None:
        query_types = accuracy_df_per_layer['query_type'].unique().tolist()

    # Filter data for specified query types
    filtered = accuracy_df_per_layer[accuracy_df_per_layer['query_type'].isin(query_types)].copy()

    if filtered.empty:
        print("No data to plot")
        return

    # Average across all labels for each query_type and layer
    avg_data = (
        filtered
        .groupby(['query_type', 'layer'], as_index=False)['acc_right']
        .mean()
        .sort_values(['query_type', 'layer'])
    )

    fig, ax = plt.subplots(figsize=figsize)

    # Plot one line per query_type
    for qt, df_qt in avg_data.groupby('query_type'):
        ax.plot(
            df_qt['layer'],
            df_qt['acc_right'],
            marker='s',
            linewidth=2,
            markersize=5,
            label=str(qt),
        )

    ax.set_xlabel('Layer', fontsize=10)
    ax.set_ylabel('Accuracy (acc_right)', fontsize=10)
    ax.set_title('All Query Types (Right Answer)', fontsize=11, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 1.05)
    ax.legend(fontsize=8, loc='best')

    # x-ticks every 5 layers
    layer_min = int(avg_data['layer'].min())
    layer_max = int(avg_data['layer'].max())
    ax.set_xticks(np.arange(layer_min, layer_max + 1, 5))

    plt.tight_layout()
    plt.show()


# Usage
plot_layer_accuracy_grid_right(accuracy_df_per_layer)
```

```python
# =========================
# 1. LINE PLOTS - Accuracy across layers (Grid Layout)
# =========================
print("\n1. Creating line plots in grid...")

def plot_layer_accuracy_grid(df, save_prefix="complex"):
    """Plot accuracy across layers for all query types in a grid layout"""
    
    # Get unique combinations of query_type and label_name
    combinations = df.groupby(['query_type', 'label_name']).size().reset_index()[['query_type', 'label_name']]
    
    if len(combinations) == 0:
        return
    
    # Calculate grid dimensions
    n_plots = len(combinations)
    n_cols = 3  # You can adjust this
    n_rows = (n_plots + n_cols - 1) // n_cols
    
    # Create figure with subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    
    # Flatten axes array for easier indexing
    if n_plots == 1:
        axes = [axes]
    else:
        axes = axes.flatten() if n_rows > 1 else axes
    
    # Plot each combination
    for idx, (_, row) in enumerate(combinations.iterrows()):
        query_type = row['query_type']
        label_name = row['label_name']
        
        # Filter data
        data = df[(df['query_type'] == query_type) & (df['label_name'] == label_name)]
        
        if len(data) == 0:
            continue
        
        ax = axes[idx]
        
        # Plot lines
        ax.plot(data['layer'], data['acc_all'], marker='o', linewidth=2, 
                label='All Inference', color='#2E86AB')
        
        if data['acc_right'].notna().any():
            ax.plot(data['layer'], data['acc_right'], marker='s', linewidth=2, 
                    label='Right Inference', color='#06A77D')
        
        if data['acc_false'].notna().any():
            ax.plot(data['layer'], data['acc_false'], marker='^', linewidth=2, 
                    label='False Inference', color='#D62246')
        
        ax.set_xlabel('Layer', fontsize=10)
        ax.set_ylabel('Accuracy', fontsize=10)
        ax.set_title(f'{query_type} - {label_name}', fontsize=11, fontweight='bold')
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)
        ax.set_ylim([0, 1])
    
    # Hide unused subplots
    for idx in range(n_plots, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Plot all query types in a grid
plot_layer_accuracy_grid(accuracy_df_per_layer, save_prefix="complex")
```

Overall Patterns:

* High variance in False Inference (red lines) - The red lines are extremely volatile and often drop to very low accuracy, suggesting the probing classifier struggles to decode information from incorrect model predictions. This makes sense - when the model fails, its internal representations may be noisy or inconsistent.
* More stable Right Inference (green lines) - Green lines are generally smoother and more stable, indicating that when the model predicts correctly, its internal representations are more consistent and decodable.
* All Inference (blue) tracks between the two - As expected, the overall accuracy falls between correct and incorrect cases.

Query Type Differences:

* Connection-based tasks (explicit_connection, explicit_arrow_direction) show dramatic differences between right/false inference, with false inference often near random
* Recognition tasks (recognition_shape, recognition_color) are more stable across all inference types
